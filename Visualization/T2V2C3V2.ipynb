```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
# distribution plot is histogram
# the line is "kernel density line"
# it is estimation, how probability
# function will look like
sns.distplot(df['age'])
```

```python
sns.distplot(df['age'], kde=False)
```

```python
sns.distplot(df['age'], kde=False, bins=2)
```

```python
# if you want to change the figsize
plt.figure(figsize=(12, 8))
sns.distplot(df['age'], kde=False, bins=50, color='green')
```
