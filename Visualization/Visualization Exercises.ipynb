```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

```python
# scatter plot price vs caret
sns.scatterplot(x='carat', y='price', data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=.1, edgecolor=None)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=.2)
```

```python
plt.figure(figsize=(10, 8))
sns.distplot(diamonds['price'], kde=False)
plt.xlim(0, 18000)
```

```python

sns.countplot(x='cut', data=diamonds)
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='cut', y='price', data=diamonds)
```

```python
diamonds['cut'].unique()
```

```python
cut_order = ['Fair', 'Ideal', 'Good', 'Very Good', 'Premium']
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='cut', y='price', data=diamonds, order=cut_order, palette='cool')
```
